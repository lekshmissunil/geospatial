Kiva reaches some of the most remote places in the world through their global network of "Field Partners". These partners are local organizations working in communities to vet borrowers, provide services, and administer loans.

In this exercise, I've investigated Kiva loans in the Philippines and identified regions that might be outside of Kiva's current network, in order to identify opportunities for recruiting new Field Partners.



```python
import geopandas as gpd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

### 1) Getting the data

The shapefile located at `loans_filepath` is loaded to create a GeoDataFrame `world_loans`.  

```python
loans_filepath = "../input/geospatial-learn-course-data/kiva_loans/kiva_loans/kiva_loans.shp"
world_loans = gpd.read_file(loans_filepath)
world_loans.head()
```

### 2) Plotting the data.

The GeoDataFrame `world` containing country boundaries is loaded.

```python
world_filepath = gpd.datasets.get_path('naturalearth_lowres')
world = gpd.read_file(world_filepath)
world.head()
```

The Kiva loan locations across the world are visualized using the `world` and `world_loans` GeoDataFrames.

```python
ax = world.plot(figsize=(20,20), color='mistyrose', linestyle=':', edgecolor='black')
world_loans.plot(ax=ax, markersize=2)
plt.show()
```

### 3) Selecting loans based in the Philippines

A GeoDataFrame `PHL_loans` which contains all rows from `world_loans` with loans that are based in the Philippines is created.

```python
PHL_loans = world_loans.loc[world_loans.country=="Philippines"].copy()
```

### 4) Understanding loans in the Philippines

A GeoDataFrame `PHL` containing boundaries for all islands in the Philippines is loaded.

```python
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw' # A KML file containing island boundaries is loaded
PHL = gpd.read_file("../input/geospatial-learn-course-data/Philippines_AL258.kml", driver='KML')
PHL.head()
```

The loans in the Philippines are visualized using the `PHL` and `PHL_loans` GeoDataFrames.

```python
ax = PHL.plot(figsize=(20,20), color='mistyrose', linestyle=':', edgecolor='black')
PHL_loans.plot(ax=ax, markersize=2)
plt.show()
```


 [This map](https://bit.ly/2U2G7x7) is used to identify any islands where it might be useful to recruit new Field Partners and any islands currently looking outside of Kiva's reach. 

There are a number of potential islands, but **Mindoro** (in the central part of the Philippines) stands out as a relatively large island without any loans in the current dataset. This island is potentially a good location for recruiting new Field Partners!
